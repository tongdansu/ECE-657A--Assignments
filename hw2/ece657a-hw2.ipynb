```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
data=pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-white.csv',header=None,names=['a','b','c','d'])
```

To record the data into DataFrame

```python
data2=data[1:11]
data3=data2[['a']]
def record(x):
    i=0
    df = pd.DataFrame(columns=[])
    for i in range (0,len(x)):
        data=x.iloc[i] 
        List=data.str.split(";").tolist()  
        df=df.append(List,ignore_index=True)
        i=i+1
    return df
data4=record(data3)
data4
```

## Task1(1) min-max normalized values
To compute the min-max normalized values by the function of each feature of the wine
And show the result in a new DataFrame

```python
def minmax(data):
    mnor = pd.DataFrame(columns=[])
    for i in range(0,10):
        df1 = pd.DataFrame(columns=[])
        df1=df1.append(data[[i]].astype('float64'))
        Min=df1.min().values
        Max=df1.max().values
        for m in range(0,len(df1)):
            k=df1.iloc[m].values
            x=(k-Min)/(Max-Min)
            mnor.loc[m,i]=x
    return mnor
minmax=minmax(data4)
minmax
```

## Task1 (2) z-score nomalized values
To compute the z-score nomalized values by the function and show the results in a new DataFrame

```python
def z_score(data):
    znor= pd.DataFrame(columns=[])
    for i in range(0,11):
        df1 = pd.DataFrame(columns=[])
        df1=df1.append(data[[i]].astype('float64'))
        Mean=df1.mean().values
        std=df1.std().values
        for m in range(0,len(df1)):
            k=df1.iloc[m].values
            x=(k-Mean)/std
            znor.loc[m,i]=x
    return znor
zscore=z_score(data4)
zscore
```

## Task1(3) mean subtracted normalized values
To compute the mean subtracted normalized values and show  the results in a new DataFrame

```python
def mean(data):
    meanor = pd.DataFrame(columns=[])
    for i in range(0,10):
        df1 = pd.DataFrame(columns=[])
        df1=df1.append(data[[i]].astype('float64'))
        Mean=df1.mean().values
        for m in range(0,len(df1)):
            k=df1.iloc[m].values
            x=(k-Mean)
            meanor.loc[m,i]=x
    return meanor
meanor=mean(data4)
meanor
```

## Task2(1) manhatten distance
To store the 2 sample datasets into 2 DataFrame
To compute the manhatten distance and show the results of both the nearest and the fasthest

```python
data5=data[1:30]
data6=data5[['a']]
data7=record(data6)
distance1= pd.DataFrame(data7.iloc[0:10])
distance2= pd.DataFrame(data7.iloc[11:21])

def manhatten_distance(data1,data2):
    mdis = pd.DataFrame(columns=["nearest",'fasthest'])
    mda = np.zeros((10, 10), dtype=np.float)
    for i in range(0,10):
        d1=data1.iloc[i].values.astype('float64')
        for m in range(0,10):
            d2=data2.iloc[m].values.astype('float64')
            md=np.sum(np.abs(d1-d2))
            mda[i][m]=md
    
    for j in range(0,10):
        mdis.loc[j,"nearest"]=mda[j].min().tolist()
        mdis.loc[j,"fasthest"]=mda[j].max().tolist()
    return mdis
mdis=manhatten_distance(distance1,distance2)
print(mdis)
```

## Task2(2) euclidean distance
To compute the euclidean distance and show the results of both the nearest and the fasthest

```python
def euclidean_distance(data1,data2):
    edis = pd.DataFrame(columns=["nearest",'fasthest'])
    eda = np.zeros((10, 10), dtype=np.float)
    for i in range(0,10):
        d1=data1.iloc[i].values.astype('float64')
        for m in range(0,10):
            d2=data2.iloc[m].values.astype('float64')
            ed=np.linalg.norm(d1-d2)
            eda[i][m]=ed
    
    for j in range(0,10):
        edis.loc[j,"nearest"]=eda[j].min().tolist()
        edis.loc[j,"fasthest"]=eda[j].max().tolist()
    return edis
edis=euclidean_distance(distance1,distance2)
print(edis)
```

## Task2(3) Cosine Distance
To compute the cosine distance and show the results of both the nearest and the fasthest

```python
def cosine_distance(data1,data2):
    cdis = pd.DataFrame(columns=["nearest",'fasthest'])
    cda = np.zeros((10, 10), dtype=np.float)
    for i in range(0,10):
        d1=data1.iloc[i].values.astype('float64')
        for m in range(0,10):
            d2=data2.iloc[m].values.astype('float64')
            cd=np.dot(d1,d2)/(np.linalg.norm(d1)*(np.linalg.norm(d2)))
            cda[i][m]=cd
    
    for j in range(0,10):
        cdis.loc[j,"nearest"]=cda[j].min().tolist()
        cdis.loc[j,"fasthest"]=cda[j].max().tolist()
    return cdis
cdis=cosine_distance(distance1,distance2)
print(cdis)
```
