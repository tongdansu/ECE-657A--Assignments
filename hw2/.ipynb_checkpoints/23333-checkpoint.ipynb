```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
data=pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-white.csv',header=None,names=['a','b','c','d'])
data2=data[1:11]
data3=data2[['a']]
def record(x):
    i=0
    df = pd.DataFrame(columns=[])
    for i in range (0,10):
        data=x.iloc[i] 
        List=data.str.split(";").tolist()  
        df=df.append(List,ignore_index=True)
        i=i+1
    return df
data4=record(data3)
def minmax(data):
    i=0
    Min=[]
    for i in range(0,11):
        df1 = pd.DataFrame(columns=[])
        df1=df1.append(data[[i]].astype('float64'))
        Min.append(df1.min())
    return Min
minmax=minmax(data4)
print(minmax)
data4
```
